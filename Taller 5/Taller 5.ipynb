```python
%pylab inline
import scipy
from scipy import misc
from scipy import ndimage
from skimage import data, color, img_as_ubyte
from skimage.feature import canny
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.draw import circle_perimeter
```

```python
#Im1 = pylab.imread("FeNod_50X_L.tif")
Im1 = pylab.imread("Nodular.jpg")
#print imagen
pylab.imshow(Im1, cmap = cm.Greys_r)
plt.title('Imagen Original', fontsize=12)
image_gray = color.rgb2gray(Im1)
```

```python
edges = canny(image_gray, sigma=3.5)
pylab.imshow(edges, cmap = cm.Greys_r)
```

```python
hough_radii = np.arange(5, 15, 1)
hough_res = hough_circle(edges, hough_radii)
accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii,
                                           total_num_peaks=50)
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10, 4))
image = color.gray2rgb(image_gray)
for center_y, center_x, radius in zip(cy, cx, radii):
    circy, circx = circle_perimeter(center_y, center_x, radius)
    image[circy, circx] = (265, 0, 0)
    
ax.imshow(image, cmap=plt.cm.gray)
plt.show()
```
