```python
%pylab inline
from scipy.optimize import curve_fit

x = np.array([3,4,5,7,8,9,11,12])
y = np.array([1.6,3.6,4.4,3.4,2.2,2.8,3.8,4.6])

xp = np.linspace(x[0], x[-1])

def func(x, a, b, c, d):
    return a*x**3+b*x**2+c*x+d

popt, pcov = curve_fit(func, x, y)

plt.plot(x, y, '*k', label='datos')
plt.plot(xp,func(xp, *popt), 'r-',
        label='fit: a=%.3f b=%.3f c=%.3f d=%.3f' 
         % (popt[0],popt[1],popt[2],popt[3]))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
